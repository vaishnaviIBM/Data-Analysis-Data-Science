```python
import requests
page = requests.get("https://www.metoffice.gov.uk/weather/forecast/gcnhtnumz#?date=2022-05-18")  # To download the web page from server
page.status_code   # Successful download of data provies a status code starrting with 2.
#page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser') 
```

```python
weathers = soup.find("ul", class_="tab-container forecast-tab-container").find_all("li")
print(weathers)
```

```python
for weather in weathers:
    Day = weather.find("div",class_="off-screen").text.split('.')[0]
    Temp = weather.find("div",class_="off-screen").text.split('.')[1]
    Condition = weather.find("div",class_="off-screen").text.split('.')[2]
    Situation = weather.find("div",class_="off-screen").text.split('.')[4]
    print(Day,Temp, Condition, Situation)
    
    
```
