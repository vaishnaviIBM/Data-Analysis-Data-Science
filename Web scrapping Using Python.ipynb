# Practice : Web Scrapping of UK weather data to analayze the data using Pandas

```python
import requests
page = requests.get("https://www.metoffice.gov.uk/services/data")  # To download the web page from server
page.status_code   # Successful download of data provies a status code starrting with 2.
#page.content
```

Using BeautifulSoup Library to parse the html page downloaded in the above cell and extract the main content of the page under p tags.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')   #create an instance of the BeautifulSoup class to parse our document
print(soup.prettify())   #Prettify to arraneg the data in a readable fashion
```
